```python
import os
import pandas as pd
import glob
import datetime

def import_data_sbs_banks(start_month, start_year, end_month=None, end_year=None):
    csv_files = glob.glob(os.path.join(r"..\..\_data\sbs\B_RawData\bancos", "*.xlsx"))
    dicc={ }
    
    if end_month is None:
        end_month=8
    if end_year is None:
        end_year=2022
        
    for f in csv_files:
        start_date=datetime.datetime.strptime(f"1/{start_month}/{start_year}","%d/%m/%Y").date()
        end_date=datetime.datetime.strptime(f"1/{end_month}/{end_year}","%d/%m/%Y").date()
        
        if (datetime.datetime.strptime((os.path.basename(f)[os.path.basename(f).find('_',6)+1:-5]).replace('_','/'),"%d/%m/%Y").date()>=start_date) and (datetime.datetime.strptime((os.path.basename(f)[os.path.basename(f).find('_',6)+1:-5]).replace('_','/'),"%d/%m/%Y").date()<=end_date):
            basename=os.path.basename(f)
            year=basename[basename.find('_',15)+1:-5]
            dicc[year]={ } 
            month=basename[basename.find('_',12)+1:-10]
            dicc[year][month]=pd.read_excel(f)

    return dicc
```
